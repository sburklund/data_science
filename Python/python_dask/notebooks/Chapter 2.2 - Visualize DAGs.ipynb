```python
import dask.delayed as delayed
from dask.distributed import Client
client = Client()

def addTwo(x):
    return x + 2

def multiplyFour(x):
    return x * 4

def sumTwoNumbers(x,y):
    return x + y

data = [1, 5, 8, 10]
data
```

```python
step1 = map(delayed(addTwo), data)
total = delayed(sum)(step1)
total.visualize()
```

```python
step2 = map(delayed(multiplyFour), step1)
total = delayed(sum)(step2)
total.visualize()
```

```python
data2 = map(lambda x: delayed(sumTwoNumbers)(x, total), data)
total2 = delayed(sum)(data2)
total2.visualize()
```

```python
totalPersisted = total.persist()
totalPersisted.visualize()
```

```python
data2 = map(lambda x: delayed(sumTwoNumbers)(x, totalPersisted), data)
total2 = delayed(sum)(data2)
total2.visualize()
```
