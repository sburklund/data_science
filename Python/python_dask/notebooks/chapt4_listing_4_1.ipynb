Listing 4.3 Finding the commn columns between the four DataFrames
- set creates list of unique column names for each dataframe
- intersection method returns a set containing the items that exist in the sets it is comparing
- reduce function pulls out the columns that are common to all 4 data frames

```python
# Import for Python 3.x
from functools import reduce

columns = [set(fy14.columns), 
          set(fy15.columns), 
          set(fy16.columns), 
          set(fy17.columns)]

common_columns = list(reduce(lambda a, i: a.intersection(i), columns))
common_columns
```

### Listing 4.6  Building a generic scheama
1. using the common_columns variable, transform each column name into a tuple with column name and np.str
2. Convert list of tuples into a dictionary

```python
import numpy as np
import pandas as pd

dtype_tuples = [(x, np.str) for x in common_columns]
dtypes = dict(dtype_tuples)
dtypes
```

### Listing 4.7  Creating a DataFrame with an explicit schema

```python
fy14 = dd.read_csv('../data/nyc-parking-tickets/Parking_Violations_Issued_-_Fiscal_Year_2014__August_2013___June_2014_.csv', dtype=dtypes)

with ProgressBar():
    display(fy14[common_columns].head())
```

### Listing 4.8  Inspecting the Vehicle Year column
- Looking at 20 of the unique values in the Vehicle Year column
    - appears we have intergers and can use *uint16* data type (unsigned integer 16 bits)

```python
with ProgressBar():
    print(fy14['Vehicle Year'].unique().head(20))
```

### Listing 4.9  Inspecting the Vehicle Year column for missing values
- using *isnull* method to check each value in the column for *np.nan*
- chaning wth *.values.any()* reduces the Boolean Series to a single *True* if at least 1 exists
    - since we have at least 1 row, cannot use *uint16*, must use *float32* datatype

```python
with ProgressBar():
    print(fy14['Vehicle Year'].isnull().values.any().compute())
```

### Listing 4.11 Applying the schema to all 4 DataFrames

- *usecols* takesa list of column names and drops any columns from the resulting DataFrame that are not in the list
